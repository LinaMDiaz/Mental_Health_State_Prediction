# Data Processing

### Vector Encoding (One Hot Encoding)

```python
# function that transform columns with 2 categorical values to dummies [0,1]
def encod(df, col):
    df.loc[:,[col]] = pd.get_dummies(df[col], drop_first=True, dtype=int).values
```

```python
def encod(df, col):
    dummy_col = pd.get_dummies(df[col], drop_first=True, dtype=int)
    df.drop(columns=[col], inplace=True)
    df[col] = dummy_col.iloc[:, 0].astype(int)  # Add dummy column back with same name
```

```python
columns = ['Veteran Status', 'Chronical diseases', 'Otherchron_group','Mental Illness','Cash Assistance Situation',
             'Unknown Insurance Coverage', 'Insured_or_Not', 'Has_Public_Insurance',
             'Has_Private_or_Other_Insurance', 'Confirmed_Medicaid_Managed'
       ]
```

```python
for col in columns:
    encod(data, col)
```

```python
data_encoded = data.copy()
data_encoded.head()
```

```python
#Encode a column with 3 or more categories using one-hot encoding.
#Drops the original column.
#Adds new dummy columns: col_value1, col_value2, etc.

def encode_multiclass_column(df, col, drop_first=True):
    dummies = pd.get_dummies(df[col], prefix=col, drop_first=drop_first, dtype=int)
    df.drop(columns=[col], inplace=True)
    df[dummies.columns] = dummies
```

```python
for column in data_encoded.columns:
    if column not in columns:
        encode_multiclass_column(data_encoded, column)
```

### Balancing data

```python
y_value=data_encoded['Mental Illness']
X1 = data_encoded.drop('Mental Illness', axis=1)  # features
```

```python
import sklearn
import imblearn
print("scikit-learn:", sklearn.__version__)
print("imblearn:", imblearn.__version__)
```

```python
from imblearn.over_sampling import SMOTE
print("Before OverSampling, counts of label '1': {}".format(sum(y_value==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_value==0)))


sm = SMOTE(random_state=2)
X_val_res, y_valor_res = sm.fit_resample(X1, y_value.ravel())
```

```python
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(X_val_res,
                                                   y_valor_res,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    stratify = y_valor_res)
```

```python
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
#here we initialize the model
MLP_model2 = MLPClassifier(hidden_layer_sizes=(20,), max_iter=500, random_state=42)
```

```python
MLP_model2.fit(X=X_train_res, y=y_train_res)
```

```python
y_test_predicted = MLP_model2.predict(X_test_res)
```

```python
from sklearn.metrics import classification_report
report_balan= classification_report(y_test_res, y_test_predicted)
print(report_balan)
```
